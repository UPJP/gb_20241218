# **2. 출력 서식 지정하기**

1. %d(정수), %f(실수), %s(문자열)를 사용하여 출력 서식을 지정
2. 서식을 제외한 문자는 입력한 그대로 출력
3. 서식을 지정해 출력할 데이터가 2개 이상일 경우 ()로 데이터를 묶어야한다

```python
print('%d' % 100)   # 정수 100이 출력
print('숫자 : %d' % 100)
print('숫자 : %d' % 100.55)

print('숫자 : %f' % 100.55)

print('문자열 : %s' % 100)
print('문자열 : %s' % 'Python')

# 5 + 3 = (5+3)
print('5 + 3 = 8')
print('%d + %d = %d' % (5,3,5+3))
```

```python
print('%5d' % 100)
print('%5d' % 10000)
print('%5d' % 1000000)
print('%05d' % 100)
```

```python
print('%6.2f' % 123.45)   # . 포함 6자리, 2자리 소숫점
print('%6.2f' % 123.456)  # 자릿수 넘어가면 반올림
print('%6.2f' % 123.4)    # 남은 자리는 0으로 채움
```

```python
print('%5s' % 'abc')
```

```python
# 5 / 3 = 1
# 5 / 3 = 1.666667
print('%d / %d = %d' % (5,3, 5/3))
print('%d / %d = %f' % (5,3, 5/3))
```

# **3. format 함수와 출력 인덱스**

```python
# 인덱스 : 0부터 순서대로 진행되는 정수
print('{0} {1}'.format('자바','학생'))
print('{0}{1}'.format('자바','학생'))
print('{1}{0}'.format('자바','학생'))
```

```python
print('{0:3s}님은 {1:03d}살 입니다'.format('김',20))
# 자바학생님의 키는 167.50cm 입니다.('자바학생',167.5)
print('{0}님의 키는 {1:6.2f}cm 입니다'.format('자바학생',167.5))
```
