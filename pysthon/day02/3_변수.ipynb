# 2. 변수의 데이터 타입

* 데이터 타입(자료형) : 자료가 저장되는 형태를 의미

```python
name = '김자바'
age = 20
weight = 50.5
isLover = False

# type() : 변수의 데이터 타입을 반환해주는 함수
print(type(name))       # <class 'str'> : 문자형 타입
print(type(age))        # <class 'int'> : 정수형 타입
print(type(weight))     # <class 'float'> : 실수형 타입
print(type(isLover))    # <class 'bool'>  : 논리형 타입
```

```python
print(isLover)
```
