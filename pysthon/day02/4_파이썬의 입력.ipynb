# 1. input 함수

* 사용자에게 데이터를 입력받을 수 있는 함수
* 데이터를 문자형으로 입력

```python
name = input()
print(name)
```

```python
print('이름을 입력하세요 : ', end='')
name = input()
print('%s님 안녕하세요!' % name)
```

```python
name = input('이름을 입력하세요 : ')
print('%s님 안녕하세요!' % name)
```

```python
# split() : 인수로 전달된 문자를 기준으로 데이터를 나눠서 저장
name, age, weight = input('이름과 나이, 몸무게를 입력하세요 : ').split(' ')
print('{}님의 나이는 {}살이며, 몸무게는 {}kg입니다.'.format(name, age, weight))
```

```python
# '/'를 구분자로 같은 내용을 입력받아오는 코드를 작성
name, age, weight = input('이름과 나이, 몸무게를 입력하세요 : ').split('/')
print('{}님의 나이는 {}살이며, 몸무게는 {}kg입니다.'.format(name, age, weight))
```

### 문제
* 국어점수, 영어점수, 수학점수를 입력받아 총점, 평균을 구하는 프로그램을 작성하자
* 단, int() : 문자열을 숫자(정수)타입으로 변환


```python
kor, eng, math = input('국어, 영어, 수학점수를 입력하세요 : ').split(' ')

tot = int(kor) + int(eng) + int(math)
avg = tot / 3

# print('총점 : ', tot)
# print('평균 : ', avg)
print('총점 : {}'.format(tot))
print('평균 : {}'.format(avg))
```
