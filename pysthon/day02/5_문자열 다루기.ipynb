# 3. 문자열 함수

* 문자열과 관련된 여러가지 기능(함수)

```python
str = '문자열 함수는 문자열과 관련된 여러가지 기능'
```

```python
# len() : 매개변수로 전달된 문자열의 개수를 반환, 파이썬 표준 함수
print(len(str))
```

```python
# count() : 문자열에서 매개변수로 전달된 문자열의 출현 횟수를 반환
print(str.count('문'))
print(str.count('문자'))
print(str.count('문장'))
```

```python
# find() : 매개변수로 전달된 문자열이 처음로 나오는 위치를 인덱스로 반환
print(str)
print(str.find('문'))
print(str.find('곽')) # 없으면 -1
```

```python
# index() : 매개변수로 전달된 문자열이 처음나오는 위치를 인덱스로 반환
print(str)
print(str.index('문'))
print(str.index('곽'))  # 없으면 에러, ValueError: substring not found
```

```python
# replace(찾을 문자열, 바꿀 문자열) : 문자열을 치환
print(str)
print(str.replace('기능','함수'))
```

```python
# upper() : 문자열을 대문자로 모두 변환
# lower() : 문자열을 소문자로 모두 변환
str = '문자열 function은 문자열과 관련된 여러가지 FUNCTION'
print(str.upper())
print(str.lower())
```

```python
str = '            Hello         Python          '
print(str)
```

```python
#lstrip() : 문자열의 왼쪽 공백을 제거
print(str.lstrip())
```

```python
#rstrip() : 문자열의 오른쪽 공백을 제거
print(str.rstrip())
```

```python
#strip() : 문자열의 양쪽 공백을 제거
print(str.strip())
```

```python
# join() : 문자열에 특정 문자열을 삽입
str = 'python'
print('/'.join(str))
```
