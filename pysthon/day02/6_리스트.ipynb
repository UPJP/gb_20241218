# 1. 리스트(List)

* 여러가지 데이터를 ,(콤마)로 구분하여 []로 감싼 자료구조

```python
li1 = [1,3,5,7]
print(li1)

li2 = list([1,3,5,7])
print(li2)

li3 = ['가가가','나나나','다다다','라라라']
print(li3)

li4 = [1, 1.5, '김자바', True]  # 타입이 다른 데이터를 저장할 수 있음
print(li4)

li5 = [1, 2, '파이썬', ['김자바', '박파이썬']]
print(li5)
```

# 2. 리스트 인덱싱

```python
li1 = [1,3,5,7]
print(li1[0])
print(li1[1])
print(li1[-1])
print(li1[0] + li1[-1])
```

```python
li2 = ['가가가','나나나','다다다',['가','나','다']]
print(li2[0])
print(li2[-1])
print(li2[-1][1])
```

```python
li3 = [1,3,5,['가가가','나나나','다다다',['가','나','다']]]
print(li3[2])
print(li3[-2])
print(li3[-1][-2])        # 다다다 출력
print(li3[-1][-1][-1])    # 다 출력
```

# 3. 리스트 슬라이싱


```python
li4 = [10,20,30,40,50]
print(li4)
print(type(li4))    # <class 'list'>
print(li4[0:3])

li5 = li4
print(li5)
li5[0] = 100;
print(li5)
print(li4)
```

```python
li6 = [10,20,30,['가가가','나나나','다다다'],40,['가','나']]
print(li6[:2])
print(li6[2:6])
print(li6[5][:1])
```

# 4. 리스트 연산자

```python
li7 = [10, 20, 30]
li8 = [40, 50, 60]

print(li7 + li8)
print(li8 + li7)

# print(li7 - li8)   # TypeError: unsupported operand type(s) for -: 'list' and 'list'

# print(li7 * li8)  # TypeError: can't multiply sequence by non-int of type 'list'

print(li7 * 3)
```

# 5. 리스트의 수정/삭제

```python
li10 = [10, 20, 30]
li10[1] = 100
print(li10)

# 슬라이싱을 이용하여 데이터를 변경한 경우 리스트에 데이터가 포함
li10[1:2] = ['가','나','다']
print(li10)

# 인덱싱을 이용하여 데이터를 변경한 경우 리스트 안에 리스트가 포함
li10 = [10, 20, 30]
li10[1] = ['가','나','다']
print(li10)

li10[1:3] = []    # 요소 삭제
print(li10)

del li10[0]
print(li10)
```
